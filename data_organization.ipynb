```python
import pandas as pd
import json
```

```python
with open('level1.json', 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data['lessons'])

df
```

```python
df['text'] = (
    "Lesson " + df['id'].astype(str) + ": " + df['title'] + "\n\n" +
    df['content'] + "\n\nExamples:\n" + df['examples'].apply(lambda x: "\n".join(x))
)

df['text'][0]
```

```python
df['text'] = df['text'].str.replace('\s+', ' ', regex=True).str.strip()
df['text'][0]
```

```python
df['topic'] = 'Python Basics'
df['difficulty'] = 'Beginner'
df['level'] = 1
```

```python
df.to_csv('level1_clean.csv', index=False)
```
