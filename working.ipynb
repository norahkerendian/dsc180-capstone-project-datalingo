```python
const items = $input.all();

const lessons = {};

for (const item of items) {
  const j = item.json;

  const lessonId = Number(j.lesson_id);
  if (!lessonId) continue;

  // Initialize lesson if not seen
  if (!lessons[lessonId]) {
    lessons[lessonId] = {
      id: lessonId,
      slug: j.lesson_slug,
      title: j.lesson_title,
      content_md: j.lesson_text || "",
      topic: j.topic,
      level: j.level,
      metadata: j.metadata || {},
      questions: []
    };
  }

  // Add questions
  const qaPairs = j.parsedQuestions?.qa_pairs || [];
  for (const qa of qaPairs) {
    lessons[lessonId].questions.push({
      id: lessons[lessonId].questions.length + 1,
      question: qa.question,
      answer: qa.answer,
      difficulty: qa.difficulty,
      category: qa.category
    });
  }
}

// Return one item per lesson
return Object.values(lessons).map(lesson => ({
  json: lesson
}));
```
